### Data Cleaning

```python
def clean_text(sentence):
    sentence = sentence.lower()
    sentence = re.sub("[^a-z]+"," ",sentence) #removing numeric from sentence
    sentence = sentence.split()
    
    sentence = [s for s in sentence if len(s)>1] #removing words of length <=1 
    sentence = " ".join(sentence)
    return sentence 
```

### Vocabulary 
- set of all unique words the model can predict

```python
descriptions = None
with open('descriptions_1.txt',"r") as f:
    descriptions = f.read()
    
json_acceptable_strings = descriptions.replace("'","\"")
descriptions = json.loads(json_acceptable_strings)
```

```python
# create our vocab
# using a set to store unique words

vocab = set()
for key in descriptions.keys():
    [vocab.update(sentence.split()) for sentence in descriptions[key]]
    # sentence is the caption corresponding to the key in dictionary
    # splitting the sentence into a list of words and then passing these words to the vocab which will store only unique words
    
print("vocab size: %d"%len(vocab))
```

```python
# total no of words across all the sentences
total_words = []

for key in descriptions.keys():
    [total_words.append(i) for des in descriptions[key] for i in des.split()]
    # des is the value in dictionary i.e. captions
    # iterating over the words a caption by splitting the caption
    
print("total words: %d"%len(total_words))
```

```python
# filter words from the vocab according to certain threshold frequency
import collections

counter = collections.Counter(total_words) # fives count of frequency of each word in total_words
freq_cnt = dict(counter)
print(freq_cnt)
```

```python
print(len(freq_cnt.keys())) # same as total no of words
```

```python
# sort this dictionary acc to the frequency count
sorted_freq_cnt = sorted(freq_cnt.items(), reverse=True, key=lambda x:x[1]) # sorting in decreasing order

# filter
threshold = 10
sorted_freq_cnt = [x for x in sorted_freq_cnt if x[1] > threshold]
total_words = [x[0] for x in sorted_freq_cnt]
```

#### STEP 1 - Image Feature Extraction

```python
# using a 50 layer deep model called ResNet. This model is already trained on imagenet dataset
# this model will allow you to extract features from images
model = ResNet50(weights="imagenet",input_shape=(224,224,3))
model.summary()
```

```python
model.layers[-2].output #last second row- gap layer
```

```python
# creating a new model out of ResNet50 conv base
model_new = Model(model.input, model.layers[-2].output)
```

```python
def preprocess_img(img):
    img = image.load_img(img,target_size=(224,224))
    img = image.img_to_array(img) # convert img into array
    img = np.expand_dims(img,axis=0) 
    # Normalisation
    img = preprocess_input(img)
    return img
```

```python
#img = preprocess_img(IMG_PATH + "1000268201_693b08cb0e.jpg")
#plt.imshow(img[0])
#plt.axis("off")
#plt.show()
#print(img)

#this is how the img will be fed to the resnet model
```

```python
# preprocessing for all images
def encode_img(img):
    img = preprocess_img(img)
    feature_vector = model_new.predict(img)
    
    feature_vector = feature_vector.reshape((-1,))
    #print(feature_vector.shape)
    return feature_vector
```

### Data pre-processing for captions
- converting captions into numbers


when you feed data to the model, every word in the sentence is a feature and that feature is
represented by a number

```python
# Vocab
len(total_words)
```

```python
word_to_idx = {} # a dict which maps words to index
idx_to_word = {} # a dict which maps index to a word

for i,word in enumerate(total_words): #iterate over all the words
    word_to_idx[word] = i+1
    idx_to_word[i+1] = word
    
```

```python
# find max length of caption to find out how many times rnn/lstm will unroll
max_len = 0
for key in train_descriptions.keys():
    for cap in train_descriptions[key]:
        max_len = max(max_len, len(cap.split()))
        
print(max_len)
```

### Data Loader/Generator

```python
def data_generator(train_descriptions,encoding_train,word_to_idx,max_len,batch_size):
    X1,X2,y = [],[],[] # batch
    
    n=0
    # loop for ever over images
    while True:
        for key,desc_list in train_descriptions.items():
            n += 1
            
            photo = encoding_train[key]
            for desc in desc_list:
                # encode the sequence (convert words into numbers)
                seq = [word_to_idx[word] for word in desc.split() if word in word_to_idx]
                for i in range(1,len(seq)):
                    xi = seq[0:i]
                    yi = seq[i]
                    
                    # padding the sequence, o denotes padding word
                    # post means adding padding after the sequence of numbers
                    xi = pad_sequences([xi],maxlen=max_len,value=0,padding='post')[0]
                    yi = to_categorical([yi],num_classes=vocab_size)[0] # one hot vector
                    
                    X1.append(photo)
                    X2.append(xi) 
                    y.append(yi) 
                    
                # yield the batch data
                if n ==batch_size:
                    yield [[np.array(X1),np.array(X2)]],np.array(y)
                    X1,X2,y = [],[],[]
                    n = 0
                
```

### Word Embeddings

```python
f = open("./glove.6B.50d.txt",encoding='utf8') # using glove embedding
```

```python
embedding_index = {} # store word vector

for line in f:
    values = line.split()
    
    word = values[0] # 1st item of list is the word, rest are embeddings
    word_embedding = np.array(values[1:],dtype='float') # embedding for a given word
    embedding_index[word] = word_embedding
```

```python
f.close()
```

```python
embedding_index['apple']
```

```python
def get_embedding_matrix():
    emb_dim = 50 # each word can be represented as 50 dimensional vector
    matrix = np.zeros((vocab_size,emb_dim)) 
    for word,idx in word_to_idx.items():
        embedding_vector = embedding_index.get(word)
        
        if embedding_vector is not None:
            matrix[idx] = embedding_vector
            
    return matrix
    
```

```python
embedding_matrix = get_embedding_matrix()
embedding_matrix.shape
```

```python
embedding_matrix[1847] # startseq and endseq are not their in glove embeddings hence 0
```

### Model Architecture

```python
# image feature extractor model
input_img_features = Input(shape=(2048,))
inp_img1 = Dropout(0.3)(input_img_features) # output of previous layer
inp_img2 = Dense(256,activation='relu')(inp_img1) # goes as input to current layer
```

```python
# partial caption sequence model
input_captions = Input(shape=(max_len,))
inp_cap1 = Embedding(input_dim=vocab_size,output_dim=50,mask_zero=True)(input_captions)
inp_cap2 = Dropout(0.3)(inp_cap1)
inp_cap3 = LSTM(256)(inp_cap2) # 256 denotes the output size(hidden vector)
```

```python
decoder1 = add([inp_img2,inp_cap3]) # adds two tensors of same dimension
decoder2 = Dense(256,activation='relu')(decoder1)
outputs = Dense(vocab_size,activation='softmax')(decoder2)

# Combined model: merge 2 networks
model = Model(inputs=[input_img_features,input_captions],outputs=outputs)
```

```python
model.summary()
```

```python
 # pre-initialise the embedding layer
model.layers[2].set_weights([embedding_matrix])
model.layers[2].trainable = False # do not train this layer i.e. Embedding Layer
```

```python
model.compile(loss='categorical_crossentropy',optimizer="adam")
```

### Training of model

```python
epochs = 20
batch_size = 3
steps = len(train_descriptions)//batch_size
```

```python
def model_train():
    
    for i in range(epochs):
        generator = data_generator(train_descriptions,encoding_train,word_to_idx,max_len,batch_size)
        model.fit_generator(generator,epochs=1,steps_per_epoch=steps,verbose=1)
        model.save('./model_weights/model_'+str(i)+'.h5')
```

```python
model_train()
```

### Predictions

```python
def predict_caption(photo):
    
    in_text = "startseq"
    for i in range(max_len):
        sequence = [word_to_idx[w] for w in in_text.split() if w in word_to_idx]
        sequence = pad_sequences([sequence],maxlen=max_len,padding='post')
        
        ypred = model.predict([photo,sequence])
        ypred = ypred.argmax() #Word with max probability always - Greedy Sampling
        word = idx_to_word[ypred]
        in_text += (' ' + word)
        
        if word == "endseq":
            break
    
    final_caption = in_text.split()[1:-1] # ignore first and last word
    final_caption = ' '.join(final_caption) # join everything with a space
    return final_caption
```
